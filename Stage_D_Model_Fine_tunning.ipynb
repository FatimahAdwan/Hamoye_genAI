```python
from datasets import load_dataset
```

```python
raw_dataset = load_dataset("glue", "mrpc")
```

Question 17

```python
from transformers import AutoTokenizer

# Load the tokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

# Tokenize the dataset with padding and truncation
encoded_dataset = raw_dataset['train'].map(lambda examples: tokenizer(examples['sentence1'], examples['sentence2'], padding='max_length', truncation=True), batched=True)

# Access the number of tokens for the 3rd sentence pair
num_tokens = len(encoded_dataset[2]['input_ids'])
print(num_tokens)
```
